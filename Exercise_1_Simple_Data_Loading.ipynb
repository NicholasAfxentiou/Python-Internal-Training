## What is a CSV File?

**CSV** stands for **Comma-Separated Values**. It's a simple way to store data in a table format.

Think of it like a spreadsheet:
- Each **row** is one observation (like one student)
- Each **column** is one variable (like "Math Score")
- Values are separated by commas

**Example CSV:**
```
Student Name,Math,Science,English
Alice,85,90,88
Bob,78,82,85
Charlie,92,88,90
```
