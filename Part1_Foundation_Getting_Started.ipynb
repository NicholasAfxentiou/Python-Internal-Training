## Section 1.2: Python Basics - Your First Steps

### Variables: Storing Information

A **variable** is like a labeled box where you store information. You give it a name and put something inside.

**Example:** Think of a variable called `age` that stores the number `25`.

Let's create our first variable:

````python
# This is a comment - Python ignores it, but it helps us understand the code
# Let's create a variable called 'age' and store the number 25
age = 25

# Now let's print (display) what's stored in 'age'
print(age)
````

**What happened?**
- `age = 25` creates a variable named `age` and stores `25` in it
- `print(age)` displays the value stored in `age`

### Basic Data Types

Python recognizes different types of data:

1. **Integers** - Whole numbers: `1`, `42`, `100`
2. **Floats** - Decimal numbers: `3.14`, `2.5`, `99.9`
3. **Strings** - Text (always in quotes): `"Hello"`, `'Python'`

Let's see examples:

````python
# Integer (whole number)
students = 30
print("Number of students:", students)
print("Type:", type(students))

# Float (decimal number)
average_score = 85.5
print("\nAverage score:", average_score)
print("Type:", type(average_score))

# String (text - notice the quotes!)
course_name = "Python for Beginners"
print("\nCourse name:", course_name)
print("Type:", type(course_name))
````

### Simple Operations

You can do math with numbers, just like a calculator:

````python
# Basic math operations
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
 print("Multiplication:", a * b)
print("Division:", a / b)

# You can also store the result in a variable
total = a + b
print("\nTotal stored in variable:", total)
````

## Section 1.3: Working with Lists

### What is a List?

A **list** is a collection of items stored together. Think of it like a shopping list!

**Example:** A list of test scores: `[85, 90, 78, 92, 88]`

### Creating Lists

Lists are created using square brackets `[]` and items are separated by commas:

````python
# Create a list of test scores
test_scores = [85, 90, 78, 92, 88]
print("Test scores:", test_scores)

# Lists can contain different types of data
student_info = ["Alice", 25, 88.5]  # Name, age, score
print("\nStudent info:", student_info)

# Lists can also contain text (strings)
subjects = ["Math", "Science", "English", "History"]
print("\nSubjects:", subjects)
````

### Accessing Items in a List

Each item in a list has a **position** (called an "index"). Important: Python starts counting at **0**, not 1!

**Index positions:**
- First item: index `0`
- Second item: index `1`
- Third item: index `2`
- And so on...

Let's see how to access items:

````python
grades = [85, 90, 78, 92, 88]

# Access the first item (index 0)
print("First grade:", grades[0])

# Access the second item (index 1)
print("Second grade:", grades[1])

# Access the last item (using -1)
print("Last grade:", grades[-1])

# Access the second-to-last item
print("Second-to-last grade:", grades[-2])
````

### Adding Items to a List

You can add new items to a list:

````python
# Start with a list
shopping_list = ["apples", "bananas", "milk"]
print("Original list:", shopping_list)

# Add a new item
shopping_list.append("bread")
print("After adding bread:", shopping_list)

# Add another item
shopping_list.append("eggs")
print("After adding eggs:", shopping_list)
````

### Why Lists Are Important

Lists are **essential** for data analysis because:
- Datasets contain many rows of data (like a list of students)
- Each row might be a list of values
- We can organize and process lots of data efficiently

## Section 1.4: Introduction to Libraries

### What are Libraries?

A **library** is a collection of pre-written code that we can use. Instead of writing everything from scratch, we can use tools that others have already created!

**Analogy:** Think of libraries like toolboxes:
- Python's basic tools = hammer, screwdriver (basic operations)
- Libraries = power tools (advanced features)

### Why Use Libraries?

- **Save time** - Don't reinvent the wheel
- **Proven code** - Tested and reliable
- **Specialized tools** - Libraries are designed for specific tasks

### Introducing Pandas

**Pandas** is a library specifically designed for data analysis. It helps us:
- Load data from files (CSV, Excel, etc.)
- Organize data in tables
- Analyze and manipulate data
- Clean and prepare data

### Installing Libraries

Before using a library, you need to install it (usually done once):

```bash
pip install pandas
```

*(If you're using Anaconda, pandas is usually already installed)*

### Importing Libraries

After installation, you need to **import** the library into your notebook. This tells Python: "Hey, I want to use pandas!"

````python
# Import pandas - this loads the library
# We use 'as pd' to give it a shorter nickname
import pandas as pd

# Now we can use pandas! Let's verify it's loaded
print("Pandas version:", pd.__version__)
print("\n✅ Pandas is ready to use!")
````

**Important Notes:**
- `import pandas as pd` - This is the standard way to import pandas
- `as pd` - This creates a shorter name so we can type `pd` instead of `pandas`
- You only need to import once per notebook (usually at the top)

### Common Libraries for Data Analysis

Here are some libraries you'll encounter:
- **pandas** (`pd`) - Data manipulation and analysis
- **numpy** (`np`) - Numerical computing
- **matplotlib** (`plt`) - Creating visualizations
- **seaborn** (`sns`) - Beautiful statistical visualizations

For now, we'll focus on **pandas** - it's the most important one for beginners!

## Section 1.5: Loading Your First Dataset

### What is a Dataset?

A **dataset** is a collection of data organized in rows and columns, like a spreadsheet or table.

**Example:** A dataset of student grades:

| Student Name | Math | Science | English |
|--------------|------|---------|----------|
| Alice        | 85   | 90      | 88       |
| Bob          | 78   | 82      | 85       |
| Charlie      | 92   | 88      | 90       |

### What is a CSV File?

**CSV** stands for **Comma-Separated Values**. It's a simple file format:
- Each row is a line in the file
- Each column is separated by a comma
- It's like Excel, but simpler and more universal

**Example CSV content:**
```
Student Name,Math,Science,English
Alice,85,90,88
Bob,78,82,85
Charlie,92,88,90
```

### Reading a CSV File

Pandas makes it easy to read CSV files. We use `pd.read_csv()`:

````python
# Make sure pandas is imported
import pandas as pd

# Load the CSV file
# Replace 'student_grades.csv' with your actual file name
df = pd.read_csv('dataset/student_grades.csv')

# 'df' stands for 'DataFrame' - pandas' way of storing tables
print("✅ Dataset loaded successfully!")
print("\nType of 'df':", type(df))
````

### What is a DataFrame?

A **DataFrame** is pandas' way of storing a table of data:
- Think of it as a super-powered spreadsheet
- It has rows (observations) and columns (variables)
- We can do lots of things with it!

### Viewing Your Data

After loading data, **always check it first!** Here's how:

````python
# Display the first 5 rows
print("First 5 rows:")
print(df.head())

# Display the first 10 rows
print("\n\nFirst 10 rows:")
print(df.head(10))

# See how many rows and columns we have
print("\n\nDataset shape (rows, columns):")
print(df.shape)

# Get basic information about the dataset
print("\n\nDataset info:")
print(df.info())
````

### Understanding the Output

**`.head()`** - Shows the first few rows (default: 5)
- Use this to see what your data looks like

**`.shape`** - Shows (number of rows, number of columns)
- Example: `(100, 5)` means 100 rows and 5 columns

**`.info()`** - Shows:
- Number of rows
- Column names
- Data types (int, float, object/string)
- How many non-null (non-empty) values

### Displaying All Columns

Sometimes datasets have many columns. To see all of them:

````python
# See all column names
print("Column names:")
print(df.columns.tolist())

# Or simply:
print("\n\nColumn names (alternative):")
print(list(df.columns))
````

### Common Issues & Solutions

**Problem:** `FileNotFoundError` - Can't find the file
- **Solution:** Make sure the CSV file is in the same folder as this notebook
- **Alternative:** Use the full path: `pd.read_csv('C:/Users/YourName/Documents/data.csv')`

**Problem:** `ModuleNotFoundError: No module named 'pandas'`
- **Solution:** Install pandas: `pip install pandas`

**Problem:** File loads but looks wrong
- **Solution:** Check the file format - make sure it's a valid CSV
- Try opening the CSV file in Excel or a text editor to verify
