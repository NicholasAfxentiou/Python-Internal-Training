## What is a Bar Chart?

A **bar chart** uses rectangular bars to show values. Each bar represents one category.

**When to use bar charts:**
- Comparing different categories (like different subjects)
- Showing counts or totals
- Comparing values across groups

**Example:** A bar chart showing test scores for different subjects

## Step 3: Prepare Data for the Bar Chart

**What do we need for a bar chart?**
- **X-axis:** Categories (like subject names: Math, Science, etc.)
- **Y-axis:** Values (like average scores)

**Let's calculate average scores for each subject:**

We'll use `.mean()` to calculate the average of each column (subject).

````python
# Step 3: Calculate average scores for each subject
# We'll select only the grade columns (not Student Name)
# Then calculate the mean (average) for each subject

# Select only the grade columns (Math, Science, English, History)
grade_columns = ['Math', 'Science', 'English', 'History']

# Calculate average for each subject
average_scores = df[grade_columns].mean()

# Display the averages
print("Average scores for each subject:")
print(average_scores)

# Store subject names and their averages
subjects = average_scores.index.tolist()  # Subject names
scores = average_scores.values.tolist()   # Average scores

print("\nSubjects:", subjects)
print("Average Scores:", scores)
````

**What happened?**
- We selected only the grade columns
- Calculated the average (mean) for each subject
- Stored the subject names and their average scores
- Now we're ready to create a bar chart!

**Understanding the code:**
- `df[grade_columns]` - selects only specific columns
- `.mean()` - calculates the average
- `.index.tolist()` - gets the column names as a list
- `.values.tolist()` - gets the average values as a list

### Key Concepts:

**Bar Chart vs Line Chart:**
- **Bar Chart:** Use for categories (subjects, months, groups)
- **Line Chart:** Use for trends over time (temperature over days)

**Basic Bar Chart Code:**
```python
plt.bar(categories, values)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Chart Title')
plt.show()
```

### Common Mistakes to Avoid:

❌ **Forgetting `plt.show()`** - Your chart won't display!  
❌ **Wrong order** - Create bars first, then add labels  
❌ **Mixing up x and y** - Categories go on x-axis, values on y-axis  
❌ **Not importing matplotlib** - Always start with `import matplotlib.pyplot as plt`
